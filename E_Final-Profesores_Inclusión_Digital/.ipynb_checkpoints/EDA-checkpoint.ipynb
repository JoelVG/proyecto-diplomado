### Realizamos una conversión de los códigos de preguntas del diccionario de datos en el formato: b1[b11] => b1.b11.

```python
# Empezamos renombrando el nombre de las columnas del diccionario de datos para un manejo más cómodo
df_dict.rename(columns={"Código": "code", "Descripción": "description"}, inplace=True)

# Realizamos la conversión de los códigos de las preguntas siguiendo el formato del df que contiene los datos
df_dict["code"] = [f.replace("[", ".").replace("]", ".") for f in df_dict["code"]]
# Cambiamos el index a la columna code
df_dict.set_index('code', inplace=True)
df_dict.head(8)
```

### Según análisis de las preguntas que se realizaron en esta encuesta se toman las más representativas para nuestro análisis.
### Las cuales se encuentran en el archivo *main_qcodes.txt*: 

```python
# Leemos los códigos de las preguntas seleccionadas, mostramos a qué preguntas corresponden
qcode_list = []
with open("main_qcodes.txt", "r") as f:
    for qc in f:
        q_code = qc.strip().replace("[",".").replace("]",".")
        question = df_dict.loc[q_code, "description"]
        qcode_list.append(q_code)
        print(f"{q_code}: {question.strip()}")
```

## Según las preguntas seleccionadas procedemos a reducir nuestro dataset de preguntas y respuestas
### Se reducen las columnas de 189 a 130. Las cuales nos darán lo necesario para este análisis

```python
df[qcode_list].head()
```

```python
# Exportamos las columnas que elegimos para el análisis
df[qcode_list].to_csv("prof_fin_clnd.csv")
```
